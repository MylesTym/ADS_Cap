```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import cleaned dataset
df = pd.read_csv('../data/clean_data.csv')
```

```python
# Dataset shape
df.info()
df.describe(include='all')
df.head()
```

```python
# Numerical metrics
num_cols = ['process_days', 'year_received', 'month_received']

for col in num_cols:

    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    sns.histplot(df[col], bins=30, kde=True)
    plt.title(f'{col} Distribution')

    plt.subplot(1, 2, 2)
    sns.boxplot(x=df[col])
    plt.title(f'{col} Boxplot')

    plt.tight_layout()
    plt.show()
```

```python
# Categorical metrics

cat_cols = ['product', 'issue', 'company', 'state']
for col in cat_cols:

    plt.figure(figsize=(10, 4))
    sns.countplot(data=df, y=col, order=df[col].value_counts().head(10).index)
    plt.title(f'Top {col} Categories')
    plt.show()
```
